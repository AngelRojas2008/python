# Listas en Python

Crear y mostrar una lista

```python
fruits = ["guanabana", "freijoa", "uva", "pitaya"]
print(fruits)
```

Acceder al primer elemento

```python
numbers = [4, 8, 15, 16, 23, 42]
print(numbers[0])
```

Acceder al último elemento

```python
numbers = [4, 8, 15, 16, 23, 42]
print(numbers[-1]) # Usar -1 es la forma más rápida de llegar al final
```

Acceder a múltiples elementos

```python
colors = ["zafre", "amaranto", "Xanadú", "Glauco"]
print(colors[1:3])
```

Longitud de una lista

```python
cities = ["Bangkok", "Bogotá", "Oslo", "Varsovia", "Ottawa"]
print(len(cities))
```

Agregar al final

```python
animals = ["ornitorrinco", "Tiburón Anguila", "Pez Bruja"]
animals.append("murcielago")
print(animals)
```

Insertar en una posición específica

```python
animals = ["ornitorrinco", "Tiburón Anguila", "Pez Bruja"]
animals.insert(1, "anguila electrica")
print(animals)
```

Eliminar por valor

```python
numbers = [10, 20, 30, 40, 50]
numbers.remove(30)
print(numbers)
```

Eliminar usando pop()

```python
letters = ["a", "b", "c", "d", "e"]
letters.pop()
print(letters)
```

Comprobar existencia (in)

```python
names = ["Ana", "Luis", "Maria", "Pedro"]
print("Maria" in names)
```

Orden ascendente (sort)

```python
numbers = [5, 2, 9, 1, 7]
numbers.sort()
print(numbers)
```

Orden descendente

```python
numbers = [5, 2, 9, 1, 7]
numbers.sort(reverse=True)
print(numbers)
```

List Comprehension (Multiplicación)

```python
numbers = [3, 6, 9, 12, 15]
multiplied = [x * 2 for x in numbers]
print(multiplied)
```

Iterar con un bucle

```python
words = ["python", "list", "exercise", "practice"]
for word in words:
    print(word)
```

Calcular el promedio

```python
grades = [80, 75, 90, 85, 95]
average = sum(grades) / len(grades)
print(average)
```
