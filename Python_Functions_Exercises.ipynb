Escribe una función llamada `average(numbers)` que reciba una lista de números y devuelva el promedio.

```python
def average(numbers):
  if not numbers:  return 0
  return sum(numbers) / len(numbers)
promedio= average([10, 20, 30, 40])
print(f"El promedio de los números es: {promedio}")
```

Escribe una función llamada `factorial(n)` que calcule el factorial de un entero positivo n.

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
factorial= factorial(5)
print(f"El factorial del numero es: {factorial}")
```
