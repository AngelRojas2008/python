Dado el diccionario:
auto = {"marca": "  ", "modelo": " ", "año": }
Imprime el valor asociado a la clave "marca".

```python
auto = {"marca": "lamborghini", "modelo": "huracan", "año": 2020}
marca = auto["marca"]
print(f"la marca es {marca}")
```

Dado el diccionario:

Dado el diccionario:
frutas = {" ": 3, " ": 5, " ": 2}
Imprime todos los valores del diccionario.

```python
frutas = {"freijoa": 3, "pitaya": 5, "guanabana": 2}
print(frutas.values())
```

Dado el diccionario:
frutas = {" ": 3, " ": 5, " ": 2}
Usa un bucle para imprimir cada clave y su valor correspondiente

```python
frutas = {"freijoa": 3, "pitaya": 5, "guanabana": 2}
for clave, valor in frutas.items():
    print(f"Clave: {clave}, Valor: {valor}")
```

Dado el diccionario:
países = {" ": " ", " ": " ", " ": " "}
Imprime cada país y su capital usando un bucle

```python
países = {"bangladesh": "Dhaka", "checoslováquia": "Prague", "republica democratica del congo": "Kinshasa"}
for pais, capital in países.items():
    print(f"La capital de {pais} es {capital}")
```

Dado el diccionario:
temperaturas = {"Lunes": 22, "Martes": 24, "Miércoles": 21, "Jueves": 23}
Encuentra e imprime la temperatura más alta del diccionario.

```python
temperaturas = {"Lunes": 22, "Martes": 24, "Miércoles": 21, "Jueves": 23}
temp_maxima = max(temperaturas.values())
print(temp_maxima)
```
